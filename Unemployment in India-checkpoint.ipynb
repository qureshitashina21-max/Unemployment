```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

### Unemployment Data Analysis

```python
df = pd.read_csv("Unemployment in India.csv")
df.head()
```

### Columns Name change

```python
df.columns = ["Region","Date",	"Frequency","Estimated Unemployment Rate","Estimated Employed","Estimated Labour Participation Rate","Area"]
```

### Null value delete and save another variable

```python
clean_df = df.dropna()
```

```python
clean_df.isnull().sum()/len(clean_df)*100
```

### Description
The dataset contains unemployment - related information collected across different 
region and time period .Key features include .

. Region - Time period of the description

. Date - Time period of the observation

. Estimated Unemployment Rate(%)-Percentage of unemployment individuals

. Estimated Labour Participation Rate (%)- workforce participation rate

. Rate - Urban or Rural classification

```python
clean_df.columns = clean_df.columns.str.strip()
```

```python
clean_df["Date"] = pd.to_datetime(
    clean_df["Date"],
    dayfirst=True,
    errors="coerce"
)
```

```python
clean_df["Months"] = clean_df["Date"].dt.month
```

```python
clean_df["Year"] = clean_df["Date"].dt.year
```

```python
clean_df.head()
```

```python
E_rate = clean_df.groupby("Year")["Estimated Unemployment Rate"].mean()
E_rate
```

### Description:
### Date se month ka naam nikala aur month-wise unemployment rate ka boxplot banaya.

```python
clean_df["Months"] = pd.to_datetime(clean_df["Date"]).dt.month_name()

plt.figure(figsize=(5,5))
sns.boxplot(x = "Months", y = "Estimated Unemployment Rate", data = clean_df,)
plt.xticks(rotation = 90)
plt.show()
```

### Particular Uttar Pradesh Unemployment Rate Months wise

```python
up_df = clean_df[clean_df["Region"] == "Uttar Pradesh"]
up_df.head()
```

```python
up_df.groupby("Months")["Estimated Unemployment Rate"].mean()
plt.figure(figsize = (4,3))
sns.barplot(x = "Months", y = "Estimated Unemployment Rate", data = up_df, hue = "Area")
plt.xticks(rotation = 90)
plt.show()
```

### Uttarakhand year wise data 

```python
Uk_df = clean_df[clean_df["Region"] == "Uttarakhand"]
#Uk_df.head()
```

```python
uk_year = Uk_df.groupby("Year")["Estimated Unemployment Rate"].mean()

plt.figure(figsize=(5,3))
plt.plot(uk_year.index, uk_year.values, marker='o')
plt.xlabel("Year")
plt.ylabel("Estimated Unemployment Rate")
plt.title("Uttarakhand  Year-wise Unemployment Rate")
plt.show()
```
